## Histogram of Popularity

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
x=data['anime_id']
xc=data['popularity']

fig,axs=plt.subplots(1,2,figsize=(12,10))
axs[0].hist(x,bins=30,color='yellow',edgecolor='black')
axs[0].set_title('Anime IDs Ranking')
axs[1].hist(xc,bins=30,color='pink',edgecolor='black')
axs[1].set_title('Popularity Rankings')
```
